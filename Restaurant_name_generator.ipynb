```python
from langchain_google_genai import GoogleGenerativeAI
model = GoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.6)
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
from langchain_core.prompts import PromptTemplate
prompt_template_name=PromptTemplate(
    template="I want to open a restaurant for {cuisine} food. Suggest a fancy name for this. only one name",
    input_variables=["cuisine"]
)


prompt_template_menu=PromptTemplate(
    template="Suggest some menu items for {restaurant_name}",
    input_variables=["restaurant_name"]
)

parser=StrOutputParser()
```

```python
chain = prompt_template_name | model | parser | prompt_template_menu | model | parser
```

```python
result=chain.invoke({"cuisine":"Pakistani"})
result
```
